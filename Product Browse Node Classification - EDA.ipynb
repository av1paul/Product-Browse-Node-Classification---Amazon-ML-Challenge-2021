```python
de=data.groupby(by='BROWSE_NODE_ID', as_index=False)['TITLE'].count().rename(columns={"TITLE": "frequency"}).sort_values('frequency',ascending=False)
```

```python
import plotly.express as px
fig = px.scatter(de, x="BROWSE_NODE_ID", y="frequency")
fig.show()
```

## Observations:

From the above graph, we can see that there are are a lot of browse node ids within the range from 0 to 0.5M where the frequency is in the a range between 1000 to 215K. This can lead to overfitting of model. We can also see that there are a lot of browse node ids where the frequency is within the range from 0 to 100. These can lead to underfitting of the model. 

## Conclusion:

Our approach after looking into this graph is that to remove all the rows of browse node ids having a frequency less than 100 and bringing all the rows of browse node ids having a frequency above 100 and selecting random 100 rows from these valid rows whose browse node id frequency is more or eqaul to 100.
